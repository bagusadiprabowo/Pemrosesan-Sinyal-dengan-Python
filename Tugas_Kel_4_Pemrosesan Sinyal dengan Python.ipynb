Pemrosesan Sinyal dengan Python


Sinyal adalah sarana yang melaluinya suatu program dapat menerima informasi dari Sistem Operasi. Ketika sistem operasi menerima kejadian tertentu, sistem operasi dapat meneruskannya ke program dalam bentuk sinyal.

dihasilkan dan diteruskan ke program.Ctrl + C pada keyboard kita, Sistem Operasi akan menghasilkan sinyal dan meneruskannya ke program. Untuk kombinasi khusus ini, sinyal SIGINT

Untuk semua sistem operasi umum, terdapat pola standar untuk menetapkan sinyal-sinyal ini, yang umumnya merupakan kependekan dari bilangan bulat.

Dalam Python, sinyal-sinyal ini didefinisikan dalam modul signal.

Sinyal Python adalah

```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft

freq = 20
amp = 1
time = np.linspace(0,2,1000)
signal1 = amp*np.sin(2*np.pi*freq*time)
plt.plot(time,signal1)
plt.xlabel('Time(s)')
plt.ylabel('Amplitudo(v)')
```

Membuat Multi Sinyal Dengan Python

```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft

#sampling rate
sr = 2000

#sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

# gelombang pertama
freq1 = 1
amp1 = 3
signal1 = amp1*np.sin(2*np.pi*freq1*t)

# gelombang kedua
freq2 = 4
amp2 = 0.5
signal2 = amp2*np.sin(2*np.pi*freq2*t)

# gelombang ketiga
freq3 = 7
amp3 = 1
signal3 = amp3*np.sin(2*np.pi*freq3*t)

# gelombang campuran/jumlah
signalAll = signal1 + signal2 + signal3

plt.figure(figsize = (32, 6))
plt.subplot(1,3,1)
plt.plot(t, signal1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1,3,2)
plt.plot(t, signal2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(1,3,3)
plt.plot(t, signal3)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
```

**Sinyal dan FFT Pada Python**

```python
plt.figure(figsize = (12, 6))
plt.subplot(1,2,1)
plt.plot (t, signalAll)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

X = fft(signalAll)
N = len(X)
n = np.arange(N)
T = N/sr
freqFFT = n/T

plt.subplot(1,2,2)
plt.stem(freqFFT, np.abs(X),'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0,10)
```
